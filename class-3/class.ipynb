```python
import math

seed = 1976123.5081254
step = 0

def sine_random(seed, step):
    num = (math.sin(seed + step)+1) * 43659.98891672314
    return num - int(num)

sine_random(seed, 0), sine_random(seed, 1), sine_random(seed, 2)
[ sine_random(seed, i) for i in range(10) ]
```

```python
from Crypto.Random import get_random_bytes
import struct

def random_int(min_val, max_val):
    range_size = max_val - min_val + 1
    random_bytes = get_random_bytes(8)  # Get 8 random bytes
    random_int = struct.unpack("Q", random_bytes)[0] % range_size  # Modulo to fit range
    return min_val + random_int

n = 100000
x = [ random_int(1, 100) for _ in range(n) ]
y = [ random_int(1, 100) for _ in range(n) ]

plt.hist2d(x, y, bins=10, cmap='inferno')
plt.colorbar()
plt.show()
```
